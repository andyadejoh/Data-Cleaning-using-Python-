```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

```

```python
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
df.info()
```

```python
missing_columns = [col for col in df.columns if df[col].isnull().any() == True]
missing_columns
```

```python
X = df.drop('Price', axis = 1)
y = df.Price
X
```

```python
Xmain = X.select_dtypes(exclude = 'object')
Xmain
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xmain, y, train_size = 0.8, random_state = 0)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
X_test.info()
```

# Dropping Columns

```python
Xmain.isnull().any()
```

```python
miss_cols = [col for col in Xmain.columns if Xmain[col].isnull().any() == True]
miss_cols
```

```python
X_train_new = X_train.drop(miss_cols, axis = 1)
X_test_new = X_test.drop(miss_cols, axis = 1)

X_test_new.info()
```

```python
from sklearn.metrics import mean_absolute_error

model = DecisionTreeRegressor()
model1 = RandomForestRegressor()
```

# Decison Tree for Dropping Columns

```python
def score (X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_preds)
    return mae
```

```python
def score1 (X_train, X_test, y_train, y_test):
    model1.fit(X_train, y_train)
    y_preds = model1.predict(X_test)
    mae = mean_absolute_error(y_test, y_preds)
    return mae
```

```python
score(X_train_new, X_test_new, y_train, y_test)
```

# Random Forest for Dropping Columns

```python
score1 (X_train_new, X_test_new, y_train, y_test)
```

# Simple Imputer (Decision Tree)

```python
imputer = SimpleImputer()
```

```python
imputed_train = imputer.fit_transform(X_train)
imputed_test = imputer.transform(X_test)

new_train = pd.DataFrame(imputed_train)
new_test = pd.DataFrame(imputed_test)

new_train.columns = X_train.columns
new_test.columns = X_test.columns

new_train.head()
```

```python
new_test.head()
```

```python
score (new_train, new_test, y_train, y_test)
```

```python
score1 (new_train, new_test, y_train, y_test)
```

# Another approach to Simple Imputer

```python
X_trn = X_train.copy()
X_tst = X_test.copy()
```

```python
for col in miss_cols:
    X_trn[col + 'was_missing'] = X_trn[col].isnull()
    X_tst[col + 'was_missing'] = X_tst[col].isnull()

    
```

```python
imputed_trn = imputer.fit_transform(X_trn)
imputed_tst = imputer.transform(X_tst)

new_trn = pd.DataFrame(imputed_trn)
new_tst = pd.DataFrame(imputed_tst)


new_trn.columns = X_trn.columns
new_tst.columns = X_tst.columns

new_trn
```

# Another approach to Simple Imputer (Decision Tree)

```python
score (new_trn, new_tst, y_train, y_test)
```

```python
score1 (new_trn, new_tst, y_train, y_test)
```

```python
# From the above simple code, 2 conclusions can be drawn namely:
# 1. The Simple Imputer method performs the best when it comes to handling missing data
# 2. The Random Forest Regression Model performs better than the Decision Tree Model (Generally Ensemble Methods perform better)
```
