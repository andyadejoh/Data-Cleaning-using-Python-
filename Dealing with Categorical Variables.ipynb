```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('melb_data.csv')
df.info()
```

```python
df.head()
```

```python
y = df.Price
X = df.drop(['Price'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
for col in X_train.columns:
    if X_train[col].isnull().any() == True:
        X_train.drop([col], axis = 1, inplace = True)
    
```

```python
for col in X_test.columns:
    if X_test[col].isnull().any() == True:
        X_test.drop([col], axis = 1, inplace = True)
    
```

```python
new_col = [col for col in X_train.columns if X_train[col].nunique()<10 and X_train[col].dtype == 'object']
neww_col = [col for col in X_train.columns if X_train[col].dtype in ['int64', 'float64']]
```

```python
coll = new_col + neww_col
X_train_full = X_train[coll].copy()
X_test_full = X_test[coll].copy()
```

```python
X_test_full
```

```python
Cat_var = []
for col in X_train_full.columns:
    if X_train_full[col].dtype == 'object' :
        Cat_var.append(col)
        
print(Cat_var)        
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def score_dataset(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor(n_estimators = 100, random_state = 0)
    model.fit(X_train, y_train)
    y_preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_preds)
    return mae
```

# Method 1 : Dropping Columns with Categorical

```python
X_train_new = X_train_full.copy()
X_test_new = X_test_full.copy()
```

```python
for thing in Cat_var:
    X_train_new.drop([thing], axis = 1, inplace = True )
```

```python
for thing in Cat_var:
    X_test_new.drop([thing], axis = 1 , inplace = True)
```

```python
X_test_new
```

```python
print('MAE for First Method(Dropping Categorical Varirables): ')
print(score_dataset(X_train_new, X_test_new, y_train, y_test))
```

# Method 2 : Label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
X_train_neww = X_train_full.copy()
X_test_neww = X_test_full.copy()
lbl = LabelEncoder()
for col in Cat_var:
    X_train_neww[col] = lbl.fit_transform(X_train[col])
    X_test_neww[col] = lbl.transform(X_test[col])
    
print("MAE from Approach 2 (Label Encoding):") 
print(score_dataset(X_train_neww, X_test_neww, y_train, y_test))    
```

# Method 3: One hot encoding

```python
X_train_OH = X_train_full.copy()
X_test_OH = X_test_full.copy()

from sklearn.preprocessing import OneHotEncoder
OHencoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
newtrain = pd.DataFrame(OHencoder.fit_transform(X_train_full[Cat_var]))
newtest = pd.DataFrame(OHencoder.fit_transform(X_test_full[Cat_var]))

newtrain.index = X_train_full.index
newtest.index = X_test_full.index

num_X_train = X_train_full.drop(Cat_var, axis=1)
num_X_valid = X_test_full.drop(Cat_var, axis=1)

OH_X_train = pd.concat([num_X_train, newtrain], axis=1)
OH_X_valid = pd.concat([num_X_valid, newtest], axis=1)


print("MAE from Approach 3 (One-Hot Encoding):") 
print(score_dataset(OH_X_train, OH_X_valid, y_train, y_test))
```
